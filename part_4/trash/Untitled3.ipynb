```python
import numpy as np
from matplotlib import pylab as plt
```

```python
x = np.arange(0, 10, 0.1)
def f(x):
    return  -100 * x ** 2 - 100 * x + 10
y = f(x)
```

```python
plt.plot(x, y)
```
