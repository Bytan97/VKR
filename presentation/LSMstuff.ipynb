```python
import numpy as np
from matplotlib import pylab as plt
```

```python
x = np.arange(0, 20, 1)
y = np.array([np.random.randint(i - 3, i + 3) for i in np.arange(0, 20, 1)])
A = np.vstack([x, np.ones(len(x))]).T
a, b = np.linalg.lstsq(A, y, rcond=None)[0]
l = a * x + b
```

```python
plt.figure()
plt.scatter(x, y, c="black")
plt.plot(x, l, "red")
plt.grid()
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.savefig("./fig/1.eps", format="eps")
plt.show()
```
